**<font size=5>wavefield sepration for q data<font size=5>**

```python
# fz data
inputfile = '/home/wangn0e/2022-2023/first_research/dataset/volve_pdown_rec_full_fmax30.npz'
inputdata = np.load(inputfile)

# Receivers
r = inputdata['r']
dr = np.mean(np.abs(np.diff(r[0])))
nr = r.shape[1]

# Sources
s = inputdata['s']
ds = np.mean(np.abs(np.diff(s[0])))
ns = s.shape[1]

# models
rho_sep = 1000      # density at receiver
vel = inputdata['vel']

# Time axis
t = inputdata['t']
ot, dt, nt = t[0], t[1]-t[0], len(t)

# # Axes
x, z = inputdata['x'], inputdata['z']
ox, dx, nx = 0.0, 50.0, len(x)
oz, dz, nz = 0.0, 20, len(z)

# Wavelet
wav, _, wav_c = ricker(t[0:51], 20)
wav_c = np.argmax(wav)

# pressure and velocities
pdown = inputdata['p']       # organized as (S, R, T)
```

**<font size=4>wavefield sepration using deghosting for pdown data at source side (nr,ns,nt)<font size=4>**

```python
# change common shot gather into common receiver gather
pdown1   = pdown.transpose(1, 0, 2)          # organized as  (S, R, T),need to be (R, S, T) 
pdown_up   = np.zeros([nr,ns,nt])
pdown_down = np.zeros([nr,ns,nt])

# velocity and density along source array
rho_s = 1000.0 
vel_s = np.mean(vel[1])

# velocity and density along receiver array
rho_r = 1200.0 
vel_r = np.mean(vel[5])

vel_dir = np.mean(vel[1:5])

z = inputdata['z'] 
dz = z[1] - z[0]

off = 0.005
clip = 2e-1  # plotting clip
```

```python
# env: UD_RM
for num_r in range(nr):
    if (num_r%10) == 0:
        print('running for receiver',num_r)
    
    p = pdown1[num_r]
    
    # Window
    
    direct = np.sqrt((r[0,num_r]-s[0])**2+(-r[1,num_r]-s[1])**2)/vel_s
    direct_off = direct + off
    win = np.ones((nt, ns))
    # iwin = np.round(direct_off / dt).astype(int)
    # for i in range(ns):
    #     win[iwin[i] :, i] = 1

    pup_s, pdown_s = pylops.waveeqprocessing.Deghosting(
        p.T,
        nt,
        ns,
        dt,
        ds,
        vel_s,
        s[1, 0]+0*dz , # 1 is better than 0.8, 1*dz
        win=win,
        npad=5,
        ntaper=11,
        solver=lsqr,
        dottest=False,
        dtype="complex128",
        **dict(damp=4e-1, iter_lim=60)
    )
    pdown_up[num_r] = pup_s.T
    pdown_down[num_r] = pdown_s.T
```

```python
# 4e-1
clip = 0.05
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(12, 9))
# fig.suptitle('deghosting before')
axs[0].imshow(pdown_down[:,ns//2].T, cmap='gray', vmin=-clip*np.abs(pdown1).max(), 
              vmax=clip*np.abs(pdown1).max(), extent=(r[0,0], r[0,-1], t[-1], t[0]))
axs[0].set_title('pdown_down'), axs[0].set_xlabel(r'$x_r$'), axs[0].set_ylabel(r'$t$')
axs[0].axis('tight')
axs[1].imshow(pdown_up[:,ns//2].T, cmap='gray', vmin=-clip*np.abs(pdown1).max(), 
              vmax=clip*np.abs(pdown1).max(), extent=(r[0,0], r[0,-1], t[-1], t[0]))
axs[1].set_title('pdown_up'), axs[1].set_xlabel(r'$x_r$'), axs[1].set_ylabel(r'$t$')
axs[1].axis('tight')
axs[1].set_ylim(3,0);
```

```python
# estimate wavelet

# wav, _, wav_c = ricker(t[0:61], 20)
# wav = wav / np.max(np.abs(np.fft.fft(wav))*dt) 
# wav_c = np.argmax(wav)

nt_wav = 100
nfft = 512 * 8
# Wavelet time axis
t_wav = np.arange(nt_wav) * (dt)
t_wav = np.concatenate((np.flipud(-t_wav[1:]), t_wav), axis=0)

# Estimate wavelet spectrum
wav_est_fft = np.mean(np.abs(np.fft.fft(pdown, nfft, axis=2)), axis=(0, 1))
fwest = np.fft.fftfreq(nfft, d=dt)
freq_taper = sp.signal.windows.tukey(nfft//16, alpha=0.1)
freq_taper[len(freq_taper)//2:] = 1
wav_est_fft[:nfft//16] *= freq_taper

# Create wavelet in time
wav_est = np.real(np.fft.ifft(wav_est_fft)[:nt_wav])
wav_est = np.concatenate((np.flipud(wav_est[1:]), wav_est), axis=0)
wav_est = wav_est / wav_est.max()
# time_taper = sp.signal.windows.tukey((nt_wav-1)*2+1, alpha=0.99)
time_taper = sp.signal.windows.kaiser(nt_wav*2-1, beta=50)
wav_est *= time_taper

wav = wav_est
wav_c = np.argmax(wav)
plt.plot(wav)
```

## remove dir from pdown_up using a win

```python
# window to remove the direct wave
toff = 0.037          # direct arrival time shift 0.05
nsmooth = 10         # time window smoothing 10

directSr_rd = flip_dirSr + toff

idirectSr_rd = np.round(directSr_rd/dt).astype(np.int64)
wrs_rd = np.ones((ns, nr, nt))

for ins in range(ns):
    for inr in range(nr):
        wrs_rd[ins, inr, :idirectSr_rd[ins, inr]]=0
```

```python
wrs_rd_rs = wrs_rd.transpose(1, 0, 2) # arranged as (R, S, T)
```

```python
# remove direct arrival
pdown_up_win = wrs_rd_rs*pdown_up
```

```python
clip = 0.05
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(12, 9))
# fig.suptitle('deghosting before')
axs[0].imshow(pdown_up[:,ns//2].T, cmap='gray', vmin=-clip*np.abs(pdown1).max(), 
              vmax=clip*np.abs(pdown1).max(), extent=(r[0,0], r[0,-1], t[-1], t[0]))
axs[0].set_title('pdown_up'), axs[0].set_xlabel(r'$x_r$'), axs[0].set_ylabel(r'$t$')
axs[0].plot(r[0], flip_dirSr[ns//2], 'r')
axs[0].plot(r[0], directSr_rd[ns//2], 'g')
axs[0].axis('tight')
axs[0].set_xlim(7000,5000)
axs[0].set_ylim(1,0)
axs[1].imshow(pdown_up_win[:,ns//2].T, cmap='gray', vmin=-clip*np.abs(pdown1).max(), 
              vmax=clip*np.abs(pdown1).max(), extent=(r[0,0], r[0,-1], t[-1], t[0]))
axs[1].set_title('pdown_up_win'), axs[1].set_xlabel(r'$x_r$'), axs[1].set_ylabel(r'$t$')
# axs[1].plot(r[0], directSr_rd[ns//2], 'g')
axs[1].axis('tight')
axs[1].set_ylim(1,0)
axs[1].set_xlim(7000,5000);
```
